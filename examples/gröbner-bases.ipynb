```julia
using Oscar, Polytropes
```

## $n=4$

```julia
G4 = complete_dag(4)
```

```julia
I4 = toric_ideal(G4)
```

```julia
GF4 = groebner_fan(I4)
```

```julia
w4 = Polytropes.interior_points_of_cones(GF4)
```

```julia
Q4 = weighted_digraph_polyhedron.(Ref(G4), w4)
```

```julia
f_vector.(Q4)
```

```julia
ow4 = weight_ordering.((v->Int.(v.//gcd(v))).(w4), Ref(base_ring(I4) |> default_ordering))
```

```julia
GB4 = map(ow4) do o
    standard_basis(I4; ordering=o)
end
```

```julia
PGB4 = begin
    PGB = MSet()
    for gb in GB4
        push!.(Ref(PGB), gb)
    end
    map(unique(PGB)) do f
        f/leading_coefficient(f)
    end |> unique
end
```

```julia
in_C = map(ow4) do o
    lt = leading_term.(unique(PGB4); ordering=o)
    map(lt) do f
        f/leading_coefficient(f)
    end |> unique
end
```
